# DATA PREPARATION OR PREPROCESSING

```python
train_datagenerator = ImageDataGenerator(
    rescale=1.0/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    rotation_range=20,
    brightness_range=[0.2,1.2],

    )
test_datagenerator = ImageDataGenerator(
    rescale=1.0/255
    )
```

```python
train_data = train_datagenerator.flow_from_directory('/content/drive/MyDrive/Tomatoes Diseases Dataset/Dataset Train',
                                                     target_size=(255,255),
                                                     batch_size=(32),
                                                     class_mode='categorical'
                                                     )

test_data = test_datagenerator.flow_from_directory('/content/drive/MyDrive/Tomatoes Diseases Dataset/Dataset Test',
                                                   target_size=(255,255),
                                                   batch_size=(32),
                                                   class_mode='categorical'
                                                   )
```
